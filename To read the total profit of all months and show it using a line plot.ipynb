```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'C:/Users/HP/Documents/Flexisaf/company_sales_data.csv'
data = pd.read_csv(url)
```

```python
data.head()
```

```python
# Exercise 1 - Read Total Profit of All Months and Show It Using a Line Plot
months = data['month_number']
total_profit = data['total_profit']

plt.figure(figsize=(10, 6))
plt.plot(months, total_profit, marker='o', linestyle='-', color='b', label='Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit per Month')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Exercise 2 - Read Data for Bathing Soap and Facewash of All Months and Display It Using Subplots
bathing_soap_sales = data['bathingsoap']
facewash_sales = data['facewash']

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Bathing Soap Sales Subplot
ax1.plot(months, bathing_soap_sales, marker='o', linestyle='-', color='g', label='Bathing Soap Sales')
ax1.set_xlabel('Month Number')
ax1.set_ylabel('Sales Units')
ax1.set_title('Bathing Soap Sales per Month')
ax1.legend()
ax1.grid(True)

# Facewash Sales Subplot
ax2.plot(months, facewash_sales, marker='o', linestyle='-', color='r', label='Facewash Sales')
ax2.set_xlabel('Month Number')
ax2.set_ylabel('Sales Units')
ax2.set_title('Facewash Sales per Month')
ax2.legend()
ax2.grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
